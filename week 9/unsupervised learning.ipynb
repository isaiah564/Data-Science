```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
sns.set()
```

```python
d=pd.read_csv('C:\\Users\\Admin\\Desktop\\DataScience\\Datasets\\zoo.csv')
d.info()
```

## agglomerative clustering

```python
labels=d['class_type']
print(np.unique(labels.values))
fig,ax=plt.subplots()
(labels.value_counts().plot(ax=ax,kind='bar'))
```

```python
d.head()
```

```python
features=d.values[:,1:-1]
features.shape
```

```python
model=AgglomerativeClustering(n_clusters=7,linkage='average',affinity='cosine')
```

```python
model.fit(features)
```

```python
model.labels_
```

```python
np.unique(model.labels_)
```

```python
labels=labels-1
```

```python
score=mean_squared_error(labels,model.labels_)
```

```python
abs_error=np.sqrt(score)
abs_error
```

```python
df=pd.read_csv('C:\\Users\\Admin\\Desktop\\DataScience\\Datasets\\driver-data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
kmeans=KMeans(2)
```

```python
df_analyze=df.drop('id',axis=1)
```

```python
kmeans.fit(df_analyze)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
len(kmeans.labels_)
```

```python
type(kmeans.labels_)
```

```python
unique,counts=np.unique(kmeans.labels_,return_counts=True)
dict(zip(unique,counts))
```

```python
df_analyze['cluster']=kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=df_analyze,hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```
