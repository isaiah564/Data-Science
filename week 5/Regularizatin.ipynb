## import relevant libaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
boston=datasets.load_boston()
```

```python
boston_d=pd.DataFrame(boston.data)
boston_d.columns=boston.feature_names
boston_d_target=np.asarray(boston.target)
boston_d['House Price']=pd.Series(boston_d_target)

x=boston_d.iloc[:,:-1]
y=boston_d.iloc[:,-1]
print(boston_d.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)
print('train the data shape of x=% s and y=% s :'%(X_train.shape,y_train.shape))
print('train the data shape of x=% s and y=% s :'%(X_test.shape,y_test.shape))
```

```python
## apply multiple linear regression model
ireg=LinearRegression()
ireg.fit(X_train,y_train)

## generate prediction on the test
ireg_pred=ireg.predict(X_test)

## calculating the mean squared error(mse)
mse=np.mean((ireg_pred-y_test)**2)
print('Mean sqaured error on test set:',mse)

### putting together the coefficient and their corresponding variable names
ireg_coef=pd.DataFrame()
ireg_coef['Columns']=X_train.columns
ireg_coef['Coefficient estimate']=pd.Series(ireg.coef_)
print(ireg_coef)
```

```python
## plotting the coefficient score
fig,ax=plt.subplots(figsize=(20,10))

color=['tab:gray','tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan','tab:orange','tab:green','tab:blue','tab:olive']
ax.bar(ireg_coef['Columns'],ireg_coef['Coefficient estimate'],color=color)
ax.spines['bottom'].set_position('zero')

plt.style.use('ggplot')
plt.show()
```

## Ridge Regression

```python
## import ridge regression from sklearn
from sklearn.linear_model import Ridge

# train the model
ridgeR=Ridge(alpha=1)
ridgeR.fit(X_train,y_train)
y_pred=ridgeR.predict(X_test)

### calculate the mean square error
mse=np.mean((y_pred-y_test)**2)
print('Mean squared error on test set:',mse)

ridge_coef=pd.DataFrame()
ridge_coef['Columns']=X_train.columns
ridge_coef['Coefficient estimate']=pd.Series(ridgeR.coef_)
print(ridge_coef)
```

```python
fig,ax=plt.subplots(figsize=(30,20))

color=['tab:gray','tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan','tab:orange','tab:green','tab:blue','tab:olive']
ax.bar(ridge_coef['Columns'],ridge_coef['Coefficient estimate'],color=color)
ax.spines['bottom'].set_position('zero')

plt.style.use('ggplot')
plt.show()
```

## lasso regression

```python
## import lasso regression
from sklearn.linear_model import Lasso

# train the model
lasso_l=Lasso(alpha=1)
lasso_l.fit(X_train,y_train)
y_pred1=lasso_l.predict(X_test)

### calculate the mean square error
mse=np.mean((y_pred1-y_test)**2)
print('Mean squared error on test set:',mse)

lasso_coef=pd.DataFrame()
lasso_coef['Columns']=X_train.columns
lasso_coef['Coefficient estimate']=pd.Series(lasso_l.coef_)
print(lasso_coef)
```

```python
fig,ax=plt.subplots(figsize=(30,20))

color=['tab:gray','tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan','tab:orange','tab:green','tab:blue','tab:olive']
ax.bar(lasso_coef['Columns'],lasso_coef['Coefficient estimate'],color=color)
ax.spines['bottom'].set_position('zero')

plt.style.use('ggplot')
plt.show()
```
